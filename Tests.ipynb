```python
import pandas as pd
import numpy as np
```

 Cat Column recognition feature from SO answer

[SO answer link](https://stackoverflow.com/questions/59384802/how-can-i-automatically-detect-if-a-colum-is-categorical/74590984#74590984)

```python
cat_cols=[]

for name, column in df.iteritems():
    unique_count = column.unique().shape[0]
    total_count = column.shape[0]
    if unique_count / total_count < 0.05:
        cat_cols.append(name)
```
